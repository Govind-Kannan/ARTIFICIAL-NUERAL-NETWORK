**import the libraries**

```python
import numpy as np
import pandas as pd
import tensorflow as tf

df=pd.read_csv('/content/lung_cancer_examples.csv')
df
```

```python
df=df.drop(['Name','Surname'],axis=1)
df
```

# Step 1: Data Preparation

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

**model creation**

**ANN**

```python
ann=tf.keras.models.Sequential()
```

```python
#first hidden layer
ann.add(tf.keras.layers.Dense(6,activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(6,activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(1,activation='sigmoid'))
```

**compile ANN**

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(x_train,y_train,batch_size=32,epochs=100)
```

```python
ann.fit(x_train, y_train, epochs=100, batch_size=32, validation_data=(x_test, y_test))
```

```python
loss, accuracy = ann.evaluate(x_test, y_test)
print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")
```

```python
# Step 6: Prediction
pred=ann.predict([[30,0,5,2]])
print(pred)
```

```python
pred=ann.predict(scaler.transform([[30,0,5,2]]))
if(pred>0.5):
  print(1)
else:
  print(0)
```
